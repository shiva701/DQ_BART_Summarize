```python
def calculate_rouge_scores(original_summaries, noisy_summaries):
    rouge = Rouge()
    rouge_scores = []
    for i in range(len(original_summaries)):
        rouge_scores.append(rouge.get_scores(original_summaries[i][0].get('summary_text'), noisy_summaries[i][0].get('summary_text')))
    return rouge_scores
```

```python
def calculate_baseline_rouge_scores(original_summaries, highlights):
    rouge = Rouge()
    rouge1_scores = []
    rouge2_scores =[]
    rougel_scores=[]
    for i in range(len(original_summaries)):
        rouge1_scores.append(rouge.get_scores(original_summaries[i][0].get('summary_text'),highlights[i] ,avg=True)['rouge-1']['f'])
        rouge2_scores.append(rouge.get_scores(original_summaries[i][0].get('summary_text'),highlights[i] ,avg=True)['rouge-2']['f'])
        rougel_scores.append(rouge.get_scores(original_summaries[i][0].get('summary_text'),highlights[i] ,avg=True)['rouge-l']['f'])
    return ("Rouge 1: ",rouge1_scores,"Rouge 2: ",rouge2_scores,"Rouge L:",rougel_scores)
```

## INV

```python
def predict_scores(input_text):
    original_summaries = [pipe(article) for article in input_text]
    noisy_summaries = [pipe(Perturb.add_random_noise(article)) for article in input_text]
    rouge_scores = []
    for i in range(len(original_summaries)):
        rouge = Rouge()
        rouge_scores.append(rouge.get_scores(original_summaries[i][0].get('summary_text'), noisy_summaries[i]))
    return rouge_scores
```

```python
threshold = 0.5
```

```python
inv_test = lambda: Editor(articles[:10]).template('The article said that {}', ['add_random_noise'])
```

```python
def threshold_prediction(input_text, threshold):
    rouge_scores = predict_scores(input_text)
    for i in range(len(rouge_scores)):
        if rouge_scores[i]['rouge-1']['f'] < threshold:
            return False
    return True
```

```python
test = INV(inv_test, name='INV Test: Add Random Noise', description='Test the model\'s ability to summarize noisy text', capability="Robustness")
```

```python
test_suite = TestSuite()
```

```python
test_suite.add(test)
```

```python
def run_with_threshold(suite, predictor, threshold):
    results = suite.run(predictor)
    for result in results:
        if result.score >= threshold:
            result.status = "PASS"
        else:
            result.status = "FAIL"
    return results
```

```python
results = run_with_threshold(test_suite, predictor=predict_scores, threshold=0.5)
```

```python
results.summary()
```

#DIV

```python
from checklist.expect import Expect
```

```python
def generate_perturbed_articles(articles, perturbations):
    perturbed_articles = []
    for article in articles:
        perturbed_articles.append([Perturb(article, p).result() for p in perturbations])
    return perturbed_articles
```

```python
from transformers.pipelines.zero_shot_image_classification import Pipeline
def predict_summary(article):
    return pipe(article)[0]['summary_text']
```

```python
def calculate_rouge_score(original_summary, predicted_summary):
    rouge = Rouge()
    scores = rouge.get_scores(original_summary, predicted_summary)
    return scores[0]['rouge-1']['f']
```

```python
editor = Editor()
p = editor.add(sentence_add=[''], sentence_swap=[('', '')], sentence_delete=['']).compile()
```

```python
orig_articles = articles[:100] # take the first 100 articles as the original articles
perturbed_articles = generate_perturbed_articles(orig_articles, p)
```

```python
test_cases = []
for i in range(len(orig_articles)):
    for j in range(len(p)):
        perturbed_summary = predict_summary(perturbed_articles[i][j])
        orig_summary = highlights[i]
        score_perturbed = calculate_rouge_score(orig_summary, perturbed_summary)
        score_original = calculate_rouge_score(orig_summary, orig_summary)
        test_cases.append({
            'perturbed_article': perturbed_articles[i][j],
            'original_summary': orig_summary,
            'perturbed_summary': perturbed_summary,
            'score_perturbed': score_perturbed,
            'score_original': score_original
        })
```

```python
test = DIR('Directionality (DIR) Test for Summarization', [
    expect_fn=Expect.(lambda x: x['score_perturbed'] >= x['score_original'],
              'Predicted summary is worse than original summary')
]))
test=TestSuite()
test.run()
```

```python
test.summary()
```

```python
def predict_summary(text):
  summary = pipe(text)
  return summary
```

```python
def generate_perturbed_articles(articles, perturbations):
    perturbed_articles = []
    for article in articles:
        perturbed_articles.append([Perturb(article, p).result() for p in perturbations])
    return perturbed_articles
```

```python
def calculate_rouge_score(original_summary, predicted_summary):
    rouge = Rouge()
    scores = rouge.get_scores(original_summary, predicted_summary)
    return scores[0]['rouge-1']['f']
```

```python
editor = Editor()
p = editor.add(sentence_add=[''], sentence_swap=[('', '')], sentence_delete=['']).compile()
```

```python
orig_articles = articles[:8] 
perturbed_articles = generate_perturbed_articles(orig_articles, p)
```

```python
test_cases = []
for i in range(len(orig_articles)):
    for j in range(len(p)):
        perturbed_summary = predict_summary(perturbed_articles[i][j])
        orig_summary = highlights[i]
        score_perturbed = calculate_rouge_score(orig_summary, perturbed_summary)
        score_original = calculate_rouge_score(orig_summary, orig_summary)
        test_cases.append({
            'perturbed_article': perturbed_articles[i][j],
            'original_summary': orig_summary,
            'perturbed_summary': perturbed_summary,
            'score_perturbed': score_perturbed,
            'score_original': score_original
        })
```

```python
t = MFT(perturbed_articles,name = 'Summarization Robustness Test', capability = "Robustness")
```

```python
test_suite = TestSuite()
test_suite.add(t)
test_suite.run()
```
